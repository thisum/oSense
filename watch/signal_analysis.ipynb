```python
import numpy as np
import pandas as pd
from scipy import pi
import matplotlib.pyplot as plt
from scipy import fftpack
from scipy import hamming
from scipy import signal
import os
%matplotlib inline
```

```python
raw_data = pd.read_csv("/Users/thisum/Documents/AppDevelopment/Java/DynamicChart/raw_data.txt")
grad_data = pd.read_csv("/Users/thisum/Documents/AppDevelopment/Java/DynamicChart/grad_data.txt")
```

```python
plt.plot(raw_data)
```

```python
f, t, Sxx = signal.spectrogram(raw_data.A, fs=25, nperseg=8, noverlap=7, mode='psd')
plt.pcolormesh(t, f, Sxx)
plt.title("8 frames")
plt.ylabel('Frequency [Hz]')
plt.xlabel('Time [sec]')
```
